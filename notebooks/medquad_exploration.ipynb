```python
from datasets import load_dataset, Dataset

# 1. Load raw dataset
raw_dataset = load_dataset("lavita/MedQuAD")["train"]
print("Raw rows:", len(raw_dataset))
print("Raw columns:", raw_dataset.column_names)

# 2. Keep only the columns needed for training
dataset = raw_dataset.remove_columns([
    col for col in raw_dataset.column_names
    if col not in ["question", "answer"]
])
print("\nAfter keeping only question/answer:", len(dataset))
print("Columns:", dataset.column_names)

# 3. Remove rows with null values
dataset = dataset.filter(
    lambda x: x["question"] is not None and x["answer"] is not None
)
print("\nAfter removing nulls:", len(dataset))

# 4. Remove rows with empty strings
dataset = dataset.filter(
    lambda x: x["question"].strip() != "" and x["answer"].strip() != ""
)
print("After removing empty strings:", len(dataset))

# 5. Normalize whitespace
def clean_text(example):
    return {
        "question": " ".join(example["question"].split()),
        "answer": " ".join(example["answer"].split())
    }

dataset = dataset.map(clean_text)
print("After whitespace cleanup:", len(dataset))

# 6. Remove only exact duplicate rows, not duplicate questions
df = dataset.to_pandas()
print("\nExact duplicate rows before removal:", df.duplicated().sum())
print("Duplicate questions (for inspection only):", df.duplicated(subset=["question"]).sum())

df = df.drop_duplicates()
dataset = Dataset.from_pandas(df, preserve_index=False)

print("\nAfter removing exact duplicate rows:", len(dataset))
print("Final columns:", dataset.column_names)

# 7. Show one sample
print("\nSample row:")
print(dataset[0])
```

```python
from datasets import load_dataset

raw_dataset = load_dataset("lavita/MedQuAD")["train"]
print("Raw rows:", len(raw_dataset))

dataset = raw_dataset.remove_columns([
    col for col in raw_dataset.column_names
    if col not in ["question", "answer"]
])
print("After keeping only question/answer:", len(dataset))

null_question = dataset.filter(lambda x: x["question"] is None)
null_answer = dataset.filter(lambda x: x["answer"] is None)
print("Rows with question=None:", len(null_question))
print("Rows with answer=None:", len(null_answer))

nonnull_dataset = dataset.filter(
    lambda x: x["question"] is not None and x["answer"] is not None
)
print("After removing nulls:", len(nonnull_dataset))

empty_question = nonnull_dataset.filter(lambda x: x["question"].strip() == "")
empty_answer = nonnull_dataset.filter(lambda x: x["answer"].strip() == "")
print("Rows with empty question:", len(empty_question))
print("Rows with empty answer:", len(empty_answer))
```
